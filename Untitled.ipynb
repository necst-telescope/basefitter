```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import TapTool, CustomJS

# 出力先ファイルを指定
output_file("tap_tool.html")

# プロットを作成
p = figure(outer_width=400, outer_height=400, tools="tap", title="Click on the plot")

# プロットにデータを追加
p.circle([1, 2, 3, 4, 5], [6, 7, 2, 4, 5], size=20)

# クリックされたときに実行されるJavaScriptコードを定義
callback = CustomJS(args=dict(p=p), code="""
    // クリックされたときの座標を取得
    var x = cb_obj['x'];
    var y = cb_obj['y'];
    console.log("Clicked at x=" + x + ", y=" + y);
""")

# TapToolを作成してコールバックを指定
tap_tool = TapTool(callback=callback)
p.add_tools(tap_tool)

# プロットを表示
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import TapTool, CustomJS, ColumnDataSource

# 出力先ファイルを指定
output_file("tap_tool.html")

# プロットを作成
p = figure(outer_width=400, outer_height=400, tools="tap", title="Click on the plot")

# プロットにデータを追加
source = ColumnDataSource(data={'x': [], 'y': []})
p.circle('x', 'y', source=source, size=20)

# クリックされたときの処理を定義する関数
def callback(attr, old, new):
    selected_index = source.selected.indices
    if selected_index:
        x = source.data['x'][selected_index[0]]
        y = source.data['y'][selected_index[0]]
        print("Clicked at x={}, y={}".format(x, y))

# ColumnDataSourceに変更を検出するコールバック関数を追加
source.on_change('selected', callback)

# プロットを表示
show(p)
```

```python
from bokeh.plotting import figure, output_file, show
from bokeh.models import TapTool, CustomJS, ColumnDataSource

# 出力先ファイルを指定
output_file("tap_tool.html")

# プロットを作成
p = figure(width=400, height=400, tools="tap", title="Click on the plot")

# プロットにデータを追加
source = ColumnDataSource(data={'x': [], 'y': []})
p.circle('x', 'y', source=source, size=20)

# クリックされたときのJavaScriptコードを定義
callback = CustomJS(args=dict(source=source), code="""
    // クリックされたときの座標を取得
    var selected_index = source.selected.indices;
    if (selected_index.length > 0) {
        var x = source.data['x'][selected_index[0]];
        var y = source.data['y'][selected_index[0]];
        console.log("Clicked at x=" + x + ", y=" + y);
    }
""")

# TapToolを作成してJavaScriptコールバックを指定
tap_tool = TapTool(callback=callback)
p.add_tools(tap_tool)

# プロットを表示
show(p)
```

```python
from random import random

from bokeh.layouts import row
from bokeh.models import ColumnDataSource, CustomJS
from bokeh.plotting import figure, output_file, show

output_file("callback.html")

x = [random() for x in range(500)]
y = [random() for y in range(500)]

s1 = ColumnDataSource(data=dict(x=x, y=y))
p1 = figure(width=400, height=400, tools="lasso_select", title="Select Here")
p1.circle('x', 'y', source=s1, alpha=0.6)

s2 = ColumnDataSource(data=dict(x=[], y=[]))
p2 = figure(width=400, height=400, x_range=(0, 1), y_range=(0, 1),
            tools="", title="Watch Here")
p2.circle('x', 'y', source=s2, alpha=0.6)

s1.selected.js_on_change('indices', CustomJS(args=dict(s1=s1, s2=s2), code="""
        var inds = cb_obj.indices;
        var d1 = s1.data;
        var d2 = s2.data;
        d2['x'] = []
        d2['y'] = []
        for (var i = 0; i < inds.length; i++) {
            d2['x'].push(d1['x'][inds[i]])
            d2['y'].push(d1['y'][inds[i]])
        }
        s2.change.emit();
    """)
)

layout = row(p1, p2)

show(layout)
```

```python
from random import random

from bokeh.layouts import row
from bokeh.models import ColumnDataSource, CustomJS
from bokeh.plotting import figure, show

x = [random() for x in range(500)]
y = [random() for y in range(500)]

s1 = ColumnDataSource(data=dict(x=x, y=y))
p1 = figure(width=400, height=400, tools="lasso_select", title="Select Here")
p1.scatter('x', 'y', source=s1, alpha=0.6)

s2 = ColumnDataSource(data=dict(x=[], y=[]))
p2 = figure(width=400, height=400, x_range=(0, 1), y_range=(0, 1),
            tools="", title="Watch Here")
p2.scatter('x', 'y', source=s2, alpha=0.6)

s1.selected.js_on_change('indices', CustomJS(args=dict(s1=s1, s2=s2), code="""
        const inds = cb_obj.indices
        const d1 = s1.data
        const x = Array.from(inds, (i) => d1.x[i])
        const y = Array.from(inds, (i) => d1.y[i])
        s2.data = {x, y}
    """),
)

layout = row(p1, p2)

show(layout)
```

```python
from bokeh.events import SelectionGeometry
from bokeh.models import ColumnDataSource, CustomJS, Quad
from bokeh.plotting import figure, show

source = ColumnDataSource(data=dict(left=[], right=[], top=[], bottom=[]))

x = [random() for x in range(500)]
y = [random() for y in range(500)]

s1 = ColumnDataSource(data=dict(x=x, y=y))

callback = CustomJS(args=dict(source=source), code="""
    const geometry = cb_obj.geometry
    const data = source.data

    // quad is forgiving if left/right or top/bottom are swappeed
    source.data = {
        left: data.left.concat([geometry.x0]),
        right: data.right.concat([geometry.x1]),
        top: data.top.concat([geometry.y0]),
        bottom: data.bottom.concat([geometry.y1])
    }
""")

p = figure(width=400, height=400, title="Select below to draw rectangles",
           tools="box_select", x_range=(0, 1), y_range=(0, 1))
p.scatter('x', 'y', source=s1, alpha=0.6)
# using Quad model directly to control (non)selection glyphs more carefully
quad = Quad(left='left', right='right',top='top', bottom='bottom',
            fill_alpha=0.3, fill_color='#009933')

p.add_glyph(source, quad, selection_glyph=quad, nonselection_glyph=quad)
print(source.data)
p.js_on_event(SelectionGeometry, callback)


show(p)
```
